```python
import plotly.express as px
df = px.data.iris()
df.head()
```

```python
fig = px.scatter(df,
                 x="sepal_width",
                 y="sepal_length",
                 color="species",
                 title="Sepal Length vs Width",
                 size='petal_length',
                 hover_data=['petal_width'])
fig.show()
```

```python
fig = px.box(df,
             x="species",
             y="petal_length",
             color="species",
             title="Petal Length Distribution by Species")
fig.show()
```

```python
fig = px.histogram(df,
                   x="sepal_length",
                   nbins=20,
                   color="species",
                   title="Histogram of Sepal Length")
fig.show()
```

```python
fig = px.parallel_coordinates(df,
                              color="species_id",
                              dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              title="Parallel Coordinates Plot of Iris Features")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load dataset
df = px.data.iris()

# Group by species and calculate mean of features
mean_df = df.groupby('species').mean(numeric_only=True).reset_index()

# Melt for tidy format
melted_df = mean_df.melt(id_vars='species', var_name='feature', value_name='average_length')

# Line plot
fig = px.line(melted_df,
              x='feature',
              y='average_length',
              color='species',
              markers=True,
              title="Average Feature Lengths per Species")
fig.show()
```

```python
fig = px.density_contour(df,
                         x="petal_width",
                         y="petal_length",
                         color="species",
                         title="Density Contour of Petal Width vs Petal Length")
fig.show()
```
