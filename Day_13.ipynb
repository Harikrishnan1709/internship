```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
planets=sns.load_dataset('planets')
```

```python
planets.head()
```

```python
print(planets.columns)
```

```python
planets['year'].value_counts()
```

```python
planets.plot(kind='scatter',x='mass',y='year')
plt.show()
```

```python
sns.pairplot(planets,hue='year')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.pairplot(planets,hue='mass')
plt.show()
```

```python
sns.distplot(planets['year'])
plt.show()
```

```python
sns.violinplot(x='year',y='mass',data=planets)
plt.show()
```

```python
sns.jointplot(x='year',y='mass',data=planets,kind='kde')
plt.show()
```
